### 可同時測試上方網頁
https://create.withcode.uk/

```python
from random import randint

while True:
    for x in range(5):
        for y in range(5):
            for i in range(10):
                display.set_pixel(x, y, i) #讓IC板亮燈，i是亮度(0-9)
                sleep(100)
                display.clear()
                
#結果會是由左到右由上而下，連續亮燈，亮度固定
```

```python
line = []
for x in range(5):
    line.append(randint(0,9))


while True:
    for i in range(5):
        display.set_pixel(2, 4-i, line(i)) 
        sleep(100)
    line.append(randint(0,9))
    line.pop(0)
```
