# HTML

```python
html_txt = "<h2>哈囉！</h2>"
egg = HTML(html_txt)
display(egg)

#<h數字>代表字體大小<h數字>
```

```python
html_txt = "<h2 style= 'color:red;'>哈囉！</h2>"
display(HTML(html_txt))

# style= 'color:red;'  更換顏色
```

```python
html_txt = "<S>哈囉！</S>"
display(HTML(html_txt))

#<s>代表刪除線<s>
```

# Python 檔案、存檔

```python
mydic = {"apple":"蘋果", "banana":"香蕉", "orange":"橘子"} 
```

```python
mydic.keys()
```

```python
for  eng  in mydic.keys():
    print(f"{eng},{mydic[eng]}")
```

```python
f= open("mydic.csv", "w")

# "w" 指存檔write
```

```python
f.close()
```

```python
%cat "mydic.csv"

# %cat 蘋果電腦的魔術指令
```

```python
!type "mydic.csv"
```

```python
f= open("mydic.csv", "r")

# "r" 指讀檔read
```

```python
raw_data =f.readlines()
```

```python
f.close()
```

```python
newdic ={}

for line in raw_data:
    eng, ch =line.rstrip('\n').split(',')
    newdic[eng]=ch
    
# rstrip(): 去掉括號中內容
```

# pickle: 直接儲存、讀取任意型態資料

```python
mydic 
```

```python
import pickle
```

```python
f = open("mydic.pickle", "wb")

# b非文字檔皆可
```

```python
pickle.dump(mydic, f)
f.close()

# 將 mydic 丟進去(存檔) f
```

## 讀回來

```python
f = open("mydic.pickle", "rb")
```

```python
pickledic = pickle.load(f)
```

```python
f.close()
```
