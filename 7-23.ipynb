# Python 檔案、存檔

```python
mydic = {"apple":"蘋果", "banana":"香蕉", "orange":"橘子"} 
```

```python
mydic.keys()
```

```python
for  eng  in mydic.keys():
    print(f"{eng},{mydic[eng]}")
```

```python
f= open("mydic.csv", "w")

# "w" 指存檔write
```

```python
f.close()
```

```python
%cat "mydic.csv"

# %cat 蘋果電腦的魔術指令
```

```python
!type "mydic.csv"
```

```python
f= open("mydic.csv", "r")

# "r" 指讀檔read
```

```python
raw_data =f.readlines()
```

```python
f.close()
```

```python
newdic ={}

for line in raw_data:
    eng, ch =line.rstrip('\n').split(',')
    newdic[eng]=ch
    
# rstrip(): 去掉括號中內容
```

# pickle: 直接儲存、讀取任意型態資料

```python
weekly_url= "https://www.marieclaire.com.tw/love-sex/astrology/43867"
```

```python
web =requests.get(weekly_url)

# requests.get()讀網頁
```

```python
weekly =BeautifulSoup(web.text,'lxml')
```

```python
weekly
```

```python
article = weekly.find_all("div",class_="articleContent")[0]
```

```python
article
```

```python
article.find_all("h2")

# find_all() 找出括號內全部內容
```

```python
article.find_all("h2")[9]

# [數字]找出第幾個
```

```python
article.find_all("h2")[9].string

# string 指只要內容
```

```python
article.find_all("h2")[9].next_sibling

# .next_sibling 下一段內容
```

```python
fortune= {}
for item in article.find_all("h2"):
    sign = (item.string[:2])
    luck = item.next_sibling.string
    fortune[sign] = luck
```
